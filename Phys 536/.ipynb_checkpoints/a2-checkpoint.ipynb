```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
plt.style.use(['science','notebook'])
```

# Question 1

First we read in the data and extract the channel, pressure, and count information:

```python
df = pd.read_csv('a2data.csv', index_col='Channel')
channels = df.index.to_numpy()
pressures = df.columns.values.astype(np.float)
counts = df.to_numpy().T
```

We can then plot the counts at different pressures

```python
plt.figure(figsize=(10,4))
[plt.plot(channels, count, label=str(pressure)+'kPa') for count,pressure in zip(counts,pressures)]
plt.legend(ncol=4)
plt.ylim(top=100)
plt.xlabel('Channel Number [arb]')
plt.ylabel('Counts')
plt.show()
```

These distributions are approximately Gaussian; We can extract the energy information by fitting each curve to a Gaussian curve and using the mean $\mu$.

Below we fit each set of data above to a Gaussian

$$Ae^{(x-\mu)/2\sigma^2}$$

and extract $A$, $\mu$, and $\sigma$ for each curve

```python
def norm(x, A, mu, sig):
    return A*np.exp(-(x-mu)**2/(2*sig**2))

def get_curve_params(count):
    x_data = channels
    y_data = count
    (A, mu, sig), _ = curve_fit(norm, x_data, y_data,
                            p0=(max(y_data), x_data[np.argmax(y_data)], 100))
    return [A, mu, sig]

As, mus, sigs = np.array([get_curve_params(count) for count in counts]).T
```

Lets confirm that the parameters were fitted properly:

```python
plt.figure(figsize=(10,4))
[plt.plot(channels, count, label=str(pressure)+'kPa') for count,pressure in zip(counts,pressures)]
[plt.axvline(mu, color='k', ls='--') for mu in mus]
plt.ylim(top=100)
plt.xlabel('Channel Number [arb]')
plt.ylabel('Counts')
plt.show()
```

The means do appear to be in the center

The uncertainty on the mean for each curve is given by

$$\sigma_{\bar{X}} = \sigma/\sqrt{n}$$

```python
ns = np.sum(counts, axis=1)
unc = sigs / ns
unc
```

These uncertainties in channels are quite small (due to the large amount of data collected) so we can be fairly confident that the mean values are quite close to the true values.

We can now plot the mean channel values explicitly vs. pressure. We will use the symbol $C_h$ to represent the channel number. Since this changes with pressure, $C_h=C_h(p)$

```python
plt.figure(figsize=(8,4))
plt.scatter(pressures, mus)
plt.xlabel('$p$ [kPa]')
plt.ylabel('$C_h$')
plt.grid()
plt.show()
```

Let $d=px$ (product of pressure and distance) be the "effective distance". We know that

* $E(d) = aC_h(d)+b$ where $E$ is the energy and $C_h(d)$ is the channel number. 
* $E(d) = \sqrt{E_0^2 - \beta d}$ where $E_0$ is the initial energy and $\beta$ is constant. This is based on the assumption that $C \propto p$ in Lecture 2 Slide 30.

Combining these we get

$$C_h(d) = \frac{1}{a}\left(\sqrt{E_0^2 - \beta d} - b \right)$$

We know the values of $px$, we know that $E_0=5.486$MeV, so we need to fit for $a$, $\beta$, and $b$.

We still need initial values for these 3 parameters. 

* The initial value of $a$ will be such that the maximum value of $\mu$ corresponds to $E_0$. 
* The initial value of $b$ will be 0. This assumes that energy and channel are directly proportional (as we will see later, this is not the case)
* The initial value for $\beta$ is such that the $\sqrt{1-x^2}$ function approximately begins and ends at the first and last data point on the plot above.



```python
d = pressures * 4.5/100
E0 = 5.486
a_init = E0 / max(mus)
b_init = 0
beta_init = E0**2 / max(d) -1 
```

The function here is channel as a function of distance.

```python
def Ch(d, a, b, beta):
    return (1/a) * (np.sqrt(np.abs(E0**2 - beta*d)) - b)
```

Use the function above to fit and obtain $a$, $b$, and $\beta$

```python
(a, b, beta), _ = curve_fit(Ch, xdata=d, ydata=mus,
                            p0=(a_init, b_init, beta_init))
```

Lets check that this is indeed a good fit

```python
x = np.linspace(min(d), max(d), 1000)
y = Ch(x, a, b, beta)
```

Plot

```python
plt.figure(figsize=(8,4))
plt.scatter(d, mus, label='Data')
plt.plot(x,y,color='r', ls='--', label='Fit')
plt.xlabel('Effective Distance [N/m]')
plt.ylabel('Channel #')
plt.legend()
plt.show()
```

We do indeed have a good fit. We can now use the parameters to extract the energies.

```python
E = a * mus + b
```

Finally, we can plot energy as a function of effective distance:

```python
plt.figure(figsize=(8,4))
plt.scatter(d, E)
plt.xlabel('Effective Distance [N/m]')
plt.ylabel('Energy [MeV]')
plt.grid()
plt.show()
```

In summary, the more air the alpha particles travel through, the more energy they lose. Furthermore, this is governed by the equation

$$E^2 = E_0^2 - Cx$$

## Part B

The Bragg peak is simply the change in energy $-dE/dx$ as the alpha particles move through the material. The negative sign is added because energy is being lost from the particle and trasnfered to the material: we are considered with energy change in the material. In our case, we have effective distance $x'=px$ so that

$$E^2 = E_0^2 - \beta x'$$

and thus

$$-\frac{dE}{dx'} = \frac{\beta}{2E} = \frac{\beta}{2\sqrt{E_0^2 - \beta x'}}$$

```python
eps = 1e-5
x = np.linspace(0, E0**2/beta - eps, 1000)
dEdx = beta / (2*np.sqrt(E0**2 - beta * x))
```

We can use this to make a plot:

```python
plt.figure(figsize=(8,4))
plt.semilogy(x, dEdx)
plt.ylabel('Energy Loss in Material $dE/dx\'$')
plt.xlabel('Effective distance $x\'=px$')
plt.grid()
plt.show()
```

Note that this function continues to rapidly increase and reach a discontinuity. This is because the forumla $E^2 =E_0^2 - Cx$ assumes slow or no change in the log term of the Bethe formula which is given by:

$$\ln(2mv^2/I_i)$$

Note that as the particle slows down, $v \to 0$. Now let $\chi = 2mv^2/I_i$. Since $\frac{d}{d\chi}\ln(\chi) = 1/\chi$, it follows that as  as $\chi \to 0$, the log term rapidly changes. Thus, in this limit, the formula $E^2 = E_0^2 - Cx$ is no longer valid. 

# Question 2

## Part A

Read data and obtain counts for each channel:

```python
df = pd.read_csv('a2data2.csv', index_col='Channel')
channels = df.index.to_numpy()
cs137 = df.Cs137
subA = df.A
subB = df.B
subC = df.C
```

Lets look at the known substance Cs137.

```python
plt.plot(channels, cs137, label='Cs137')
plt.legend()
plt.show()
```

The mean energy of the central peak is at 662keV; this comes from the decay of $^{156m}Ba$ to its stable state during the overall decay process. 

Assuming that $E \propto C_h$, we can calibrate the data by finding the mean value of the central peak

```python
channels_sub = channels[(channels>725)*(channels<850)]
cs137_sub = cs137[(channels>725)*(channels<850)]
```

We can fit this to a Gaussian curve to extract the mean

```python
def norm(x, A, mu, sig):
    return A*np.exp(-(x-mu)**2/(2*sig**2))

(A, mu, sig), _ = curve_fit(norm, xdata=channels_sub, ydata=cs137_sub,
                            p0=(max(cs137_sub), channels_sub[np.argmax(cs137_sub)], 70))
```

Lets make sure the fit is good:

```python
x = np.linspace(min(channels_sub), max(channels_sub), 1000)
y = norm(x, A, mu, sig)
plt.plot(channels_sub, cs137_sub, label='Cs137')
plt.plot(x, y, label='Best Fit')
plt.plot()
plt.legend()
plt.show()
```

Looks good! Now since $E = \alpha C_h$ where $\alpha$ is a proportionality constant we get

$$\alpha = E_{peak}/C_{h,peak}$$

```python
E_peak = 662
alpha = E_peak / mu
```

Now we can use alpha to get our energies

```python
energies = alpha * channels
```

```python
fig, ax = plt.subplots(1,3,figsize=(13,4))
ax[0].plot(energies, subA, label='A')
ax[1].plot(energies, subB, label='B')
ax[2].plot(energies, subC, label='C')
ax[1].set_xlabel('Energies [keV]')
ax[0].set_ylabel('Counts')
ax[0].set_title('Substance A')
ax[1].set_title('Substance B')
ax[2].set_title('Substance C')
fig.tight_layout()
plt.show()
```

We know that these correspond to Na22, Mn54, and Co60. All information used below was obtained from:

https://www-nds.iaea.org/relnsd/vcharthtml/VChartHTML.html

**The Photopeaks**

* It is well known that Na22 undergoes $\beta^{+}$ decay which eventually results in electron positron annihilation. During this annihilation, two photons with energies of 511keV are emitted. As such, since substance C has a peak at approximately 511keV, is is likely that **Substance C is Na22**. In addition, the peak at approximately 1274keV comes about as follows: after a $\beta^{+}$ decay, Na22 becomes an excited isomer of Ne22, which eventually decays to the ground state of Ne22 through the emission of a 1274keV photon.

* Co60 decays through $\beta^{-}$ emission to various isomers of Ni60. The main decay involves a decay to one excited isomer, followed by the emission of a 1173keV photon to a less-excited isomer, followed by yet another photon emission of 1333keV to reach the ground state of Ni60. The peaks seen on substance A are at these energies, and thus it follows that **Substance A is Co60**.

* Mn54 decays through electron capture or Beta+ decay to an excited isomer of Cr54, which decays to its ground state through the emission of an 835keV photon. As such, it follows that **Substance B is Mn54**.

**Compton Edges**

There are a few other interesting features to note, besides just the photopeaks. The region under 1000keV, 700keV, and 400keV for substances A, B, and C respectively correspond to compton edges. This can be confirmed as follows: the maximum energy transfered to an electron during compton scattering is

$$E_{e^{-},\text{max}} = E \left( 1 - \frac{1}{1+\frac{2E}{m_e c^2}} \right)$$

This corresponds to a cutoff point for compton interaction: we should see no counts at energies higher than this that originate from compton scattering. We can compute this quantity for all 3 substances (for each photo peak) and see where these lines lie.

**Back Scatter Peaks**:

The angular distribution in the Klein-Nishima formula has a local maximum at $\theta=180^{\circ}$. Since the sources used emit photons isotrpoically, some of the photons (which are initiall travelling in a direction opposite to the detector) are backscattered at $\theta \approx 180^{\circ}$ (now travelling towards the detector). The detector then picks up these photons, which have an energy of $E=E_{\text{photopeak}}-E_{\text{compton edge}}$. These peaks are also seen on the graph.

## Compton Edges

We manually plot the compton edges on the data

```python
peakC1 = 511
peakC2 = 1274
peakA1 = 1173
peakA2 = 1333
peakB1 = 835

def get_Emax(E):
    return E*(1-1/(1+2*E/511))

Ee_A1 = get_Emax(peakA1)
Ee_A2 = get_Emax(peakA2)
Ee_B1 = get_Emax(peakB1)
Ee_C1 = get_Emax(peakC1)
Ee_C2 = get_Emax(peakC2)
```

Plot

```python
fig, ax = plt.subplots(1,3,figsize=(13,4))
ax[0].plot(energies, subA, label='A')
ax[0].axvline(Ee_A1, ls='--', color='k')
ax[0].axvline(Ee_A2, ls='--', color='k')
ax[1].plot(energies, subB, label='B')
ax[1].axvline(Ee_B1, ls='--', color='k')
ax[2].plot(energies, subC, label='C')
ax[2].axvline(Ee_C1, ls='--', color='k')
ax[2].axvline(Ee_C2, ls='--', color='k')
ax[1].set_xlabel('Energies [keV]')
ax[0].set_ylabel('Counts')
ax[0].set_title('Substance A')
ax[1].set_title('Substance B')
ax[2].set_title('Substance C')
fig.tight_layout()
plt.show()
```

While the lines computed clearly represent compton edges in substances B and C, it is less easy to see in substance A. The first compton edge, corresponding to the photopeak at 1173keV is certainly easy to see, but since the compton edge of the 1333keV photopeak overlaps with the 1173 photopeak, it is not visible in the plot.

The distribution of counts before the compton edges are not uniform; this is due to the the fact that the **(i)** the angle of the scattered photon determines the likelihood of interaction (Klein-Nishima) and **(ii)** the energy of the emitted electron depends on the angle of photon scattering.

## Back Scatter Peaks

Get the energy of the back scatter peaks

```python
Eb_A1 = peakA1 - get_Emax(peakA1)
Eb_A2 = peakA2 - get_Emax(peakA2)
Eb_B1 = peakB1 - get_Emax(peakB1)
Eb_C1 = peakC1 - get_Emax(peakC1)
Eb_C2 = peakC2 - get_Emax(peakC2)
```

Plot

```python
fig, ax = plt.subplots(1,3,figsize=(13,4))
ax[0].plot(energies, subA, label='A')
ax[0].axvline(Eb_A1, ls='--', color='k')
ax[0].axvline(Eb_A2, ls='--', color='k')
ax[1].plot(energies, subB, label='B')
ax[1].axvline(Eb_B1, ls='--', color='k')
ax[2].plot(energies, subC, label='C')
ax[2].axvline(Eb_C1, ls='--', color='k')
ax[2].axvline(Eb_C2, ls='--', color='k')
ax[1].set_xlabel('Energies [keV]')
ax[0].set_ylabel('Counts')
ax[0].set_title('Substance A')
ax[1].set_title('Substance B')
ax[2].set_title('Substance C')
fig.tight_layout()
plt.show()
```

As shown above, all back scattered peaks are in the expected locations.
